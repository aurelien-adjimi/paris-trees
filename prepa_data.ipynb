# Analyse et Visualisation du Nombre d’Arbres par Arrondissement à Paris

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Lire le fichier CSV avec le point-virgule comme séparateur
dataset = pd.read_csv('p2-arbres-fr.csv', sep=';')

# Vérifier les valeurs manquantes
print(dataset.isnull().sum())

# Supprimer les valeurs manquantes dans les colonnes 'arrondissement' et 'circonference_cm'
dataset = dataset.dropna(subset=['arrondissement', 'circonference_cm'])

# Supprimer les doublons
dataset = dataset.drop_duplicates()

# Calculer le nombre d'arbres par arrondissement
nombre_arbres = dataset['arrondissement'].value_counts()

# Afficher le nombre d'arbres par arrondissement
print(nombre_arbres)

# Tracer un graphique du nombre d'arbres par arrondissement
plt.figure(figsize=(10,6))
sns.barplot(x=nombre_arbres.index, y=nombre_arbres.values)
plt.title('Nombre d\'arbres par arrondissement')
plt.xticks(rotation=90)
plt.show()
```

# Visualisation du Nombre d’Arbres Remarquables par Arrondissement à Paris




```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Lire le fichier CSV avec le point-virgule comme séparateur
dataset = pd.read_csv('p2-arbres-fr.csv', sep=';')

# Remplacer les valeurs manquantes par NaN
dataset = dataset.replace(r'^\s*$', np.nan, regex=True)

# Supprimer les valeurs manquantes dans les colonnes 'arrondissement' et 'remarquable'
dataset = dataset.dropna(subset=['arrondissement', 'remarquable'])

# Supprimer les doublons
dataset = dataset.drop_duplicates()

# Filtrer pour ne garder que les arbres remarquables
dataset_remarquable = dataset[dataset['remarquable'] == 1]

# Tracer un graphique du nombre d'arbres remarquables par arrondissement
plt.figure(figsize=(10,6))
sns.countplot(data=dataset_remarquable, x='arrondissement')
plt.title('Nombre d\'arbres remarquables par arrondissement')
plt.xticks(rotation=90)
plt.show()


```

# Visualisation du Nombre d’Arbres par Espèce pour les 10 Espèces les Plus Courantes à Paris




```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Lire le fichier CSV avec le point-virgule comme séparateur
dataset = pd.read_csv('p2-arbres-fr.csv', sep=';')

# Remplacer les valeurs manquantes par NaN
dataset = dataset.replace(r'^\s*$', np.nan, regex=True)

# Supprimer les valeurs manquantes dans la colonne 'espece'
dataset = dataset.dropna(subset=['espece'])

# Supprimer les doublons
dataset = dataset.drop_duplicates()

# Créer un sous-ensemble du DataFrame avec seulement les 10 espèces les plus courantes
top_especes = dataset['espece'].value_counts().index[:10]
dataset_top_especes = dataset[dataset['espece'].isin(top_especes)]

# Tracer un graphique du nombre d'arbres par espèce
plt.figure(figsize=(10,6))
sns.countplot(data=dataset_top_especes, y='espece', order=top_especes)
plt.title('Nombre d\'arbres par espèce')
plt.show()
```

# Visualisation du Nombre d’Arbres par Genre et Espèce pour les 10 Genres et Espèces les Plus Courantes à Paris




```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Lire le fichier CSV avec le point-virgule comme séparateur
dataset = pd.read_csv('p2-arbres-fr.csv', sep=';')

# Remplacer les valeurs manquantes par NaN
dataset = dataset.replace(r'^\s*$', np.nan, regex=True)

# Supprimer les valeurs manquantes dans les colonnes 'genre' et 'espece'
dataset = dataset.dropna(subset=['genre', 'espece'])

# Supprimer les doublons
dataset = dataset.drop_duplicates()

# Créer une nouvelle colonne 'genre_espece' qui combine le genre et l'espèce
dataset['genre_espece'] = dataset['genre'] + " " + dataset['espece']

# Créer un sous-ensemble du DataFrame avec seulement les 10 genres et espèces les plus courants
top_genre_espece = dataset['genre_espece'].value_counts().index[:10]
dataset_top_genre_espece = dataset[dataset['genre_espece'].isin(top_genre_espece)]

# Tracer un graphique du nombre d'arbres par genre et espèce
plt.figure(figsize=(10,6))
sns.countplot(data=dataset_top_genre_espece, y='genre_espece', order=top_genre_espece)
plt.title('Nombre d\'arbres par genre et espèce')
plt.show()

```

# Visualisation de la Distribution Géographique des Arbres à Paris


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Read the CSV file with semicolon as the separator
dataset = pd.read_csv('p2-arbres-fr.csv', sep=';')

# Replace missing values with NaN
dataset = dataset.replace(r'^\s*$', np.nan, regex=True)

# Drop missing values in the 'geo_point_2d_a' and 'geo_point_2d_b' columns
dataset = dataset.dropna(subset=['geo_point_2d_a', 'geo_point_2d_b'])

# Convert the geo_point columns to numeric
dataset['geo_point_2d_a'] = pd.to_numeric(dataset['geo_point_2d_a'])
dataset['geo_point_2d_b'] = pd.to_numeric(dataset['geo_point_2d_b'])

# Create a scatter plot of the trees' locations
plt.figure(figsize=(10,6))
sns.scatterplot(data=dataset, x='geo_point_2d_b', y='geo_point_2d_a')
plt.title('Geographical distribution of trees in Paris')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```

# Visualisation de la Distribution de la Circonférence des Arbres à Paris (moins de 200 cm)




```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Lire le fichier CSV avec le point-virgule comme séparateur
dataset = pd.read_csv('p2-arbres-fr.csv', sep=';')

# Remplacer les valeurs manquantes par NaN
dataset = dataset.replace(r'^\s*$', np.nan, regex=True)

# Supprimer les valeurs manquantes dans la colonne 'circonference_cm'
dataset = dataset.dropna(subset=['circonference_cm'])

# Convertir la colonne 'circonference_cm' en numérique
dataset['circonference_cm'] = pd.to_numeric(dataset['circonference_cm'])

# Filtrer les données pour ne garder que les arbres dont la circonférence est inférieure à 200 cm
dataset_filtered = dataset[dataset['circonference_cm'] < 200]

# Créer un histogramme de la circonférence des arbres
plt.figure(figsize=(10,6))
sns.histplot(data=dataset_filtered, x='circonference_cm', bins=30)
plt.title('Distribution de la circonférence des arbres à Paris (moins de 200 cm)')
plt.xlabel('Circonférence (cm)')
plt.ylabel('Nombre d\'arbres')
plt.show()
```

# Visualisation Cartographique du Nombre d’Arbres par Arrondissement à Paris




```python
import pandas as pd
import folium

# Lire le fichier CSV avec le point-virgule comme séparateur
dataset = pd.read_csv('p2-arbres-fr.csv', sep=';')

# Calculer le nombre d'arbres par arrondissement
trees_per_arr = dataset['arrondissement'].value_counts()

# Créer une carte centrée sur Paris
m = folium.Map(location=[48.8566, 2.3522], zoom_start=12)

# Pour chaque arrondissement, ajouter un marqueur à la carte
for arr in trees_per_arr.index:
    # Calculer la position moyenne des arbres dans cet arrondissement
    mean_lat = dataset[dataset['arrondissement'] == arr]['geo_point_2d_a'].mean()
    mean_lon = dataset[dataset['arrondissement'] == arr]['geo_point_2d_b'].mean()
    
    # Ajouter un marqueur à la carte
    folium.Marker(
        location=[mean_lat, mean_lon],
        popup=f"{arr}: {trees_per_arr[arr]} arbres",
    ).add_to(m)

# Afficher la carte
m


```
